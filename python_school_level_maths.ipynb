##Expanding brackets
The expand function can be used to help you check your algebra or to perform complicated expansions. Make sure you put a * in between any brackets to multiply them. For example you can expand the following  
$(x+1)(x+2)(x+3)(x+4)$  
$(x+3)(x+5)^3$   
$(x+3)^3$

```
out[1] = s.expand((x+1)*(x+2)*(x+3)*(x+4))
out[2] = s.expand((x+3)*(x+5)**2)
out[3] = s.expand((x+3)**3)
pprint(1,2,3)

```

##Simplifying expressions
The simplify function can be used to simplify algebraic expressions. For example you can simplify  
$x^4+x^3+x^2-3x^4+4x^3$

```
out[4] = s.simplify(x**4 +x**3 + x**2 -3*x**4 +4*x**3)
pprint(4)
```

##Solving equations
We can use python to solve equations. For example we can solver the equation $x^2 - 1 =0$. To do this we use the solve function. In the example below. Notice how the =0 is missed out. This is because the solve function can only solve equations that are equal to zero. Notice also that we need to tell the solve function what the variable in the equation is by putting in a comma followed by the x.

```
out[6] = s.solve(x**2 - 1, x)
pprint(6)
```

##Solving simultaneous equations
This is very similar to solving one variable equations. All equations must be of the form $=0$ and we must tell the solve function what the different variables are. For example, we can solve   
$x+5y=2$   
$-3x+6y=15$   
First we have to rearrange them to make them $=0$   
$x+5y-2=0$   
$-3x+67-15=0$   

```
out[14] = s.solve((x + 5*y - 2, -3*x + 6*y - 15), x, y)
pprint(14)
```

##Trigonometric Functions
You can find values of sin, cos and tan or the inverse of them by putting an a on the front, acos, asin and atan. Note how we are using the math library m.) when we are dealing with numerical calcualtion. f we want to do algebraic calculations we need to use the sympy library (s.)

```
out[15] = m.sin(m.pi/2)
out[16] = m.cos(3*m.pi/2)
out[17] = m.tan(m.pi/4)
out[18] = m.asin(0)
out[19] = m.acos(1)
out[20] = m.atan(m.pi)
pprint(15,15,17,18,19,20)
```

##Differentiation
You can find derivatives. For example  
$\frac{d}{dx}\cos^2(x)$  
$\frac{d^3}{dx^3}x^4$

```
out[21] = s.diff((s.cos(x))**2, x)
out[22] = s.diff(x**4, x, 3)
pprint(21,22)
```

##Integration
You can also integrate. Note how we make the infinity symbol oo. For example   
$\int \cos(x) dx$  
$\int_0^{\infty}e^x dx$    


```
out[23] = s.integrate(s.cos(x), x)
out[24] = s.integrate(s.exp(-x), (x, 0, s.oo))
pprint(23,24)
```

##Vectors and matrices
You can define matrices and vectors. Multiply them, find inverses and determinants. For example if  
A = $\begin{pmatrix}
1 & -1 \\
3 & 4 \\
0 & 2
\end{pmatrix}
B = \begin{pmatrix}
1 & 2 & 4 \\
3 & 4 & -2
\end{pmatrix}
C = \begin{pmatrix}
1 & -1 \\
3 & 4
\end{pmatrix}$  
Find $A \cdot B, \det{B},$ and $C^{-1}$

```
A = s.Matrix([[1, -1], [3, 4], [0, 2]])
B = s.Matrix([[1, 2, 4], [3, 4, -2]])
C = s.Matrix([[1, -1], [3, 4]])
out[25] = A*B
out[26] = C.det()
out[27] = C**-1
pprint(25,26,27)
```

##Repeating things using loops
The great thing about a programming language is you can get it to repeat things multiple times really easily. For example, lets say for example, we want to expand all of the expressions $(1+x)^n$ for n between 1 and 10, we could do it like this:

```
results = []
for n in range(1,11):
  expansion = s.expand((x+3)**n)
  results.append(expansion)
pprint(results)
  
```

##Units
We can make calcualtions using different units by using a library called Pint. For example, lets say we want to calculate a speed from a distance in kilometers and a time in minutes, but we want the answer in meters per second. There are hundreds of units including anything you are likely to ever want in science. A full list an be found [here](https://github.com/hgrecco/pint/blob/master/pint/default_en.txt). 

```
d = 24.0 * units.kilometers
t = 3 * units.minutes
speed = d/t
speed.ito(units.meters / units.seconds)

print(speed)
```

##Prefixes
We can deal with prefixes simply by adding in to the unit defintinition. All of the standard scientific units are [available](https://github.com/hgrecco/pint/blob/master/pint/default_en.txt). For example, the above example modified with micro-meters and yocto-seconds with a conversion to meters per second.

```
d = 3.5 * units.micrometers
t = 2.1 * units.yoctoseconds
speed = d/t
speed.ito(units.meters / units.seconds)
print(speed)

```

##Error Bounds
Lets say we have the equation for gravitation force between two masses  
$F_g = \frac{Gm_1m_2}{r}$  
and we and we have the values for
$G,m_1, m_2$ and $r$ defined as follows with error bounds given  
$m_1 = 3.5 \pm 0.1$ mega kilogrammes  
$m_1 = 2.1 \pm 0.2$   mega kilogrammes  
$r = 73000 \pm  1000$  
Lets say we also want to convert the answer into base units.

```
m1 = ufloat(3.5, 0.1) * units.megakilogram
m2 = ufloat(2.1, 0.2) * units.megakilogram
r = ufloat(73000,1000) * units.kilometer
G= sc.constants.G
F = G*m1*m2/r
print(F.to_base_units())
      
```
